# The simple Excel simulator
Javier has calibrated everything to deal with integers, here is how the excel looks like
![](2022-08-17-09-34-27.png)

```python
%matplotlib inline
# Load dependencies
from pycel import ExcelCompiler
import pycel
import matplotlib.pyplot as plt
import os
import pandas as pd

filename = r'simple_simulator.xlsx'
```

```python
#
# This needs to be serialized and loaded directly to avoid
# parsing of the excel every time the simulator is loaded 
# (Delete the .pkl file if you want to modify the excel)
#

# load & compile the file to a graph
excel = ExcelCompiler(filename=filename,cycles=True)

# We need to tell the excel compiler which cells we are interested in
supply_shock = excel.evaluate('simulator!J6')
inflation_sequence = excel.evaluate('simulator!A1:E30')

# Now we can save the graph to a pickle file
print("Serializing to disk...")
excel.to_file(filename + '.pkl')
```

```python
# We can now display the parsed excel
display(supply_shock)
display(inflation_sequence)
```

```python
# Show it in a nice dataframe
pd.DataFrame(inflation_sequence[1:], columns=inflation_sequence[0])
```

# Simulator

```python
# for some reason we need to reload the compiler
excel = ExcelCompiler.from_file(filename + '.pkl')

# Change the shock size parameter and reevaluate the excel
excel.set_value('simulator!J6', -0.04) # Change to -8%
display(excel.evaluate('simulator!J6'))
inflation_sequence = excel.evaluate('simulator!A1:E30')

# Show new results in a nice dataframe
pd.DataFrame(inflation_sequence[1:], columns=inflation_sequence[0])
```
